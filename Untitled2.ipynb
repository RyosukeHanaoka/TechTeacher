```python
#pydicomを呼び出し、さらにテスト用データを呼び出す。
import pydicom
from pydicom.data import get_testdata_file#'get_testdata_file'は、テストデータ呼び出し用のモジュール
filename = get_testdata_file("rtplan.dcm")#"rtplan.dcm"というテストデータを呼び出して、変数"filename"に代入する。
ds = pydicom.dcmread(filename)#"filename"のメタデータを呼び出して、変数"ds"に代入する。
```

```python
ds#変数dsを表示
```

```python
ds[0x20, 0x0e].value
```

```python
#データエレメントのキーワード、あるいはタグ番号を使用してデータを呼び出す
ds.PatientName
ds[0x10,0x10].value#タグ番号を使用して呼び出す場合、４桁のタグ番号の２桁目をxにする。
```

```python
#エレメントのキーワードあるいはtagナンバーを使用して、エレメントの値を代入することもできる。
ds.PatientID = "12345"
ds.SeriesNumber = 5
ds[0x10,0x10].value = 'Test'
```

```python
ds.BeamSequence[0].BeamName
# Or, set an intermediate variable to a dataset in the list
beam1 = ds.BeamSequence[0]  # First dataset in the sequence
beam1.BeamName
```

```python
#正確なタグ番号やキーワードがわからなくても大丈夫。キーワードに含まれる文字列を、dir()の引数に入れると、その文字列を含むキーワードを探してくれる。
ds.dir("pat")
```

```python
from pydicom import dcmread
from pydicom.data import get_testdata_file
filename = get_testdata_file("MR_small.dcm")
ds = dcmread(filename)
ds.PixelData
```

```python
ds.pixel_array
```

```python
ds.pixel_array.shape
```

```python
# example: zero anything < 300
arr = ds.pixel_array
arr[arr < 300] = 0
ds.PixelData = arr.tobytes()
ds.save_as("temp.dcm")
```

```python
from pydicom.pixel_data_handlers.util import apply_modality_lut

fname = get_testdata_file("CT_small.dcm")
ds = dcmread(fname)
arr = ds.pixel_array
hu = apply_modality_lut(arr, ds)
```

```python
from pydicom.pixel_data_handlers.util import apply_voi_lut

fname = get_testdata_file("MR-SIEMENS-DICOM-WithOverlays.dcm")
ds = dcmread(fname)
arr = ds.pixel_array
out = apply_voi_lut(arr, ds, index=0)
```

```python
#DICOM画像を表示する方法１：matplotlibの使用
import matplotlib.pyplot as plt
import pydicom
from pydicom.data import get_testdata_files
filename = get_testdata_files("CT_small.dcm")[0]
ds = pydicom.dcmread(filename)
plt.imshow(ds.pixel_array, cmap=plt.cm.bone)
```
